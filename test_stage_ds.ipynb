# 2. Modelization

> One hot encoding for the categorical data

For each column, any category having less than 10 data points should be tagged as "other". This way the number of categories can be reduced by a huge amount. We will have less dummy columns.

```python
def dataEncoding(df,data,column):
    columnStats = data.groupby(column)[column].agg('count')
    # find the variables having less than 10 data points
    columnStatsLessThan10 = columnStats[columnStats<=10]
    # replace them by 'other'
    data[column] = data[column].apply(lambda x:'other' if x in columnStatsLessThan10 else x)
    # one hot encoding
    dummiesColumn = pd.get_dummies(data[column])
    
    # concat and drop the dummy column of 'other' in order to not have the "Dummy Variable Trap" 
    # (in order to have independant variables, in other words so that the 'other' dummy column couldn't be predicted from the other variables)
    if df.empty :
        if 'other' in dummiesColumn:
            df = pd.concat([data,dummiesColumn.drop('other',axis='columns')],axis='columns')
        else :
            df = pd.concat([data,dummiesColumn],axis='columns')
            
    else :
        if 'other' in dummiesColumn:
            df = pd.concat([df,dummiesColumn.drop('other',axis='columns')],axis='columns')
        else:
            df = pd.concat([df,dummiesColumn],axis='columns')
            
            
    df = df.drop(column,axis='columns')
    # we drop the column once the one hot encoding is done
    
    return df
```

```python
df = pd.DataFrame()

# one hot encoding
df = dataEncoding(df,data,'ProductCode')
df = dataEncoding(df,data,'SpecialHandlingCodeList')
df = dataEncoding(df,data,'CommodityCode')
df = dataEncoding(df,data,'POS')
df = dataEncoding(df,data,'FlownMonth')

# drop uncessary variables
df = df.drop(['DocumentRatingSource','FlownYear',
              'AgentCode','AgentName','POSCountryName',
              'OriginCode','DestinationCode','CargoType'],axis='columns')
```

> Building the model

```python
# features and target
X = df.drop(['Revenue','RevenuePerKg'],axis='columns')
y= df.RevenuePerKg
```

First we will work with a simple train test split and then we will work with cross validation to test the ability of the model to predict new data.

```python
from sklearn.model_selection import train_test_split
# random_state = 10
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

I chose to work with Random Forest Regressor because it is based on ensemble learning so basically if my trees are giving a result above 0.5 and i'm using enough trees, according to the law of large numbers the model will necessarly converge to a decent result.

The importat step is to tune the hyperparameters so that the model don't overfit or underfit. But I have to do this manually by testing some parameters (and ofc knowing what will happen when we increase for example a certain parameter) because I don't have a the choice of doing a grid search since it needs a high computational power.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
regr = RandomForestRegressor(n_estimators=100,max_depth=100, random_state=0)
regr.fit(X_train, y_train)

print('R2 score on the train set : ',regr.score(X_train,y_train))
print('R2 score on the test set : ',regr.score(X_test,y_test))
```

```python
y_pred = regr.predict(X_test)
fig1 = plt.figure()
plt.scatter(y_test,y_pred)
plt.xlabel('y_test')
plt.ylabel('y_pred')
plt.title('Comparaison of the real price per kg and the predicted price per kg')
plt.plot(y_test,y_test,'r')
plt.show()
```

We see that the predicted price is pretty close to real one with the presence of some errors that we can avoid by introducing conditions when the model is in production ( for example for a specified distance and specified type we know the range of prices that can take place )

To show the importance of cross validation we will repeat this same process but with a different random state for the 'train test split' so we will have a different shuffling for the data so a different train set and a different test set

```python
# random_state = 5
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=5)
regr = RandomForestRegressor(n_estimators=100,max_depth=100, random_state=0)
regr.fit(X_train, y_train)

print('R2 score on the train set : ',regr.score(X_train,y_train))
print('R2 score on the test set : ',regr.score(X_test,y_test))
```

So we use k fold cross validation to see how the model will perform in general

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(RandomForestRegressor(n_estimators=100,max_depth=100, random_state=0), X, y, cv=cv)
```

Overall the model performs well. The closer the R2 score to 1 the better the regression fit. On the 5 folds only one time the model performed below 0.9 for the R2 score.
